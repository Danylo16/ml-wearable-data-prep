```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/session_01.csv")
df.head()
```

```python
for col in ["ax", "ay", "az", "heart_rate", "skin_temp"]:
    df[col].plot(figsize=(12,3), title=col)
    plt.show()

df.groupby("activity")["ax"].plot(legend=True, figsize=(10,3))
plt.show()
```

```python
feat = pd.read_csv("../data/features/session_01_features.csv")

feat.boxplot(
    column=["ax_mean", "ay_mean", "az_mean", "sma"],
    by="activity",
    figsize=(10,6)
)
plt.suptitle("")
plt.show()
```
